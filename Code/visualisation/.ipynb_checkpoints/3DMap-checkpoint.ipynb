```python
%matplotlib widget
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import Normalize
import ipywidgets as widgets
from mpl_toolkits.mplot3d import Axes3D
```

```python
files_to_compare = 1
start_slice = 49
end_slice = 51
```

```python
# list for the images
image_list = []

# list for names
names_list = []

# create input fields for the files
hBoxes = []

for i in range (files_to_compare):
    input_image = widgets.Text(placeholder='Select image file (.mhd)')
    image_list.append(input_image)
    input_name = widgets.Text(placeholder='Select name')
    names_list.append(input_name)

    hbox = widgets.HBox([input_image, input_name])
    hBoxes.append(hbox)

for i in range (files_to_compare):
    display(hBoxes[i])    

# display boxes
def read_input_values():
    image_values = [widget.value for widget in image_list]
    name_values = [widget.value for widget in names_list]
    return image_values, name_values

# button click event handler
def on_button_click(b):
    global image_list, names_list
    image_list, names_list = read_input_values()

read_button = widgets.Button(description='Read Inputs')
read_button.on_click(on_button_click)

# display the button
display(read_button)
```

```python
# load images
image_list = [image.replace('"', '') for image in image_list]
print(image_list)
images = [sitk.GetArrayFromImage(sitk.ReadImage(file)) for file in image_list]
```

```python
def plot_3d_slice_with_axis_shift(ax, image, slice_index, title="3D Surface Plot"):

    slice_data = image[slice_index, :, :]  

    if slice_data.size == 0:
        raise ValueError("The extracted slice is empty. Check the slice index or the input data.")

    # Create a grid for the y and z dimensions
    y = np.arange(slice_data.shape[0])
    z = np.arange(slice_data.shape[1])

    # Define axis transformations
    y_offset = -((slice_data.shape[0] / 2) * 500) - 250
    z_offset = -((slice_data.shape[1] / 2) * 500) - 250
    y_step = 30  # Scale for y-axis
    z_step = 30  # Scale for z-axis

    # Apply scaling and offset to the axes
    Y, Z = np.meshgrid(y * y_step + y_offset, z * z_step + z_offset)

    # Create the 3D surface plot
    surface = ax.plot_surface(
        Y,
        Z,
        slice_data.T,
        cmap='viridis',
        edgecolor='none',
        alpha=0.8
    )

    # Set plot titles and labels
    ax.set_title(title, fontsize=10)
    ax.set_xlabel('Distance (um)', fontsize=8)
    ax.set_ylabel('Distance (um)', fontsize=8)
    ax.set_zlabel('Intensity (Gbq)', fontsize=8)

    return surface



num_images = len(images)
num_slices = end_slice-start_slice + 1

fig, axes = plt.subplots(num_images, num_slices, figsize=(18, 6 * num_images), subplot_kw={'projection': '3d'})

if num_images == 1:
    axes = np.expand_dims(axes, axis=0)  

for i in range(num_images):  
    for j, slice_index in enumerate(range(start_slice, end_slice+1)):  
        ax = axes[i, j]
        plot_3d_slice_with_axis_shift(ax, images[i], slice_index=slice_index, title=f'Image {i + 1}, Slice {slice_index}')

plt.tight_layout()
plt.show()
```
